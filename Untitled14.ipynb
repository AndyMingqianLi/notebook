```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

Simple vs. Complex Model: You are comparing the performance of a simple model (Attack + Defense) and a more complex model that incorporates interactions between several variables.

In-Sample vs. Out-of-Sample Evaluation: The R-squared values provide an understanding of how well the model fits the training data (in-sample) and how well it generalizes to new data (out-of-sample).

Model Complexity: The complex model with many predictors and interactions may overfit the training data if it is too flexible. Evaluating both in-sample and out-of-sample R-squared helps assess whether this complexity improves performance or if the model is just memorizing the data.

Impact of Centering and Scaling on Multicollinearity
The code performs centering and scaling of the predictors in both models (model3_center_scale_fit and model4_CS_fit) to address multicollinearity:

Centering: Subtracts the mean of the predictor variable so that it has a mean of zero.
Scaling: Divides by the standard deviation to standardize the predictor variables.
In model3, centering and scaling reduce the condition number from 343.0 to 1.66, which indicates a significant improvement in the numerical stability of the model. However, when centering and scaling are applied to model4 (the more complex model), the condition number increases drastically to 2.25 trillion, indicating that multicollinearity still exists.

Multicollinearity contributes to poor "out of sample" generalization because it leads to:

Overfitting: The model might fit the training data too closely, capturing noise or irrelevant patterns that don't generalize well to new, unseen data.
Instability: The estimated regression coefficients become unstable, meaning small changes in the data can lead to large changes in the coefficients.
High Variance in Predictions: The model's predictions may vary significantly with slight changes in the input data, making it unreliable for out-of-sample predictions.
