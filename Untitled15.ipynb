Q1

A Classification Decision Tree is a type of supervised learning algorithm used to solve classification problems. These are problems where the goal is to predict a categorical outcome based on input features. The tree systematically splits the dataset into branches using decision rules derived from feature values, aiming to maximize the separation of classes at each node.

Examples of Real-World Applications:
Medical Diagnosis: Predicting whether a patient has a particular disease (e.g., diabetes: "Yes" or "No") based on diagnostic tests.
Fraud Detection: Classifying transactions as "fraudulent" or "non-fraudulent" in financial systems.
Customer Segmentation: Identifying customer categories (e.g., "loyal customer," "new customer," or "at-risk customer") in marketing.
Spam Email Filtering: Determining whether an email is "spam" or "not spam."
Loan Approval: Assessing whether a loan applicant is "approved" or "not approved" based on credit history and other factors.

Difference in Prediction Methods
1. Classification Decision Tree:
How It Makes Predictions:
It uses a series of binary splits based on input features, forming a tree structure.
At each leaf node (end of a branch), the tree assigns the class label that is most frequent among the data points that fall into that node.
For example, if 70% of samples in a leaf are "Yes" and 30% are "No," the tree will predict "Yes" for any new data point reaching that leaf.
Predictions are categorical.
2. Multiple Linear Regression:
How It Makes Predictions:
It models the relationship between one dependent variable (numerical) and one or more independent variables (numerical or categorical) by fitting a linear equation.


Key Differences Between Classification Decision Tree and Multiple Linear Regression
Type of Output:
A Classification Decision Tree predicts categorical outcomes, such as "Yes" or "No." In contrast, Multiple Linear Regression predicts continuous numerical values, such as a house price of $350,000.

Nature of the Model:
A Classification Decision Tree is a non-parametric model, meaning it does not make assumptions about the data's underlying distribution. It uses a rule-based, tree-like structure for predictions. On the other hand, Multiple Linear Regression is a parametric model, which assumes a linear relationship between the dependent and independent variables and fits a linear equation.

Prediction Logic:
A Classification Decision Tree predicts the most frequent class at a leaf node after systematically splitting data based on feature values. Multiple Linear Regression directly calculates a numerical prediction by substituting feature values into a fitted linear equation.

Use Cases:
Classification Decision Trees are well-suited for classification tasks, such as spam email detection or customer segmentation. In contrast, Multiple Linear Regression is used for regression problems, like predicting house prices or temperature changes.

Q2

Accuracy
Scenario: Best used in balanced datasets where both positive and negative outcomes are equally important.
Example: Diagnosing general diseases in a population where the prevalence is evenly distributed.
Rationale: Accuracy provides an overall view of the classifier's performance. However, it can be misleading in imbalanced datasets, as it does not distinguish between false negatives and false positives.


Sensitivity
Scenario: Critical in applications where missing a positive case has severe consequences.
Example: Screening for life-threatening diseases like cancer or detecting fraud in transactions.
Rationale: A high sensitivity ensures that most of the actual positive cases are identified, reducing the chances of false negatives, which are critical in such scenarios.


Specificity
Scenario: Important when correctly identifying negatives is crucial, such as avoiding unnecessary treatments.
Example: Confirming that healthy individuals are not misdiagnosed in a medical test (e.g., avoiding false alarms in radiology for tumors).
Rationale: A high specificity ensures fewer false positives, which is vital to prevent unnecessary interventions or alarm.


Precision
Scenario: Useful in contexts where the cost of false positives is high.
Example: Email spam filters, where mislabeling an important email as spam has a significant cost, or recommending critical decisions like approving loans.
Rationale: High precision minimizes false positives, ensuring that positive predictions are reliable and actionable.

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, recall_score, make_scorer
import graphviz as gv

url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")
# create `ab_reduced_noNaN` based on the specs above
del ab['Weight_oz']
del ab['Width']
del ab['Height']
       
ab_dropna = ab.dropna()
       
       
new_data_types = {'Hard_or_Paper': "category", 
                  'NumPages': int,
                  'Pub year': int}

ab_reduced_noNaN = ab_dropna.astype(new_data_types)
```

```python
from sklearn.model_selection import train_test_split

# Split the data
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Report the sizes of the training and testing datasets
print("Number of observations in the training set:", len(ab_reduced_noNaN_train))
print("Number of observations in the test set:", len(ab_reduced_noNaN_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Define the target and feature
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']  # Binary target variable (1 for Hardcover, 0 for Paperback)
X = ab_reduced_noNaN[['List Price']]  # Feature: List Price

# Step 1: Instantiate the model with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Step 2: Train the model
clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(10, 8))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Define the target and features
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']  # Target: 1 for Hardcover, 0 for Paperback
X = ab_reduced_noNaN[['NumPages', 'Thick', 'List Price']]  # Features: NumPages, Thick, List Price

# Create the DecisionTreeClassifier with max_depth=4
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)

# Train the classifier
clf2.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(15, 10))
tree.plot_tree(
    clf2, 
    feature_names=['NumPages', 'Thick', 'List Price'], 
    class_names=['Paperback', 'Hardcover'], 
    filled=True,
    fontsize=10
)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, recall_score

# Define the test set features and target for clf
X_test_clf = ab_reduced_noNaN_test[['List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # 1 for Hardcover, 0 for Paperback

# Predictions for clf
y_pred_clf = clf.predict(X_test_clf)

# Define the test set features for clf2
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

# Predictions for clf2
y_pred_clf2 = clf2.predict(X_test_clf2)

# Function to calculate sensitivity, specificity, and accuracy
def evaluate_model(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()  # Extract values from confusion matrix

    sensitivity = tp / (tp + fn)  # Recall for positive class
    specificity = tn / (tn + fp)  # Recall for negative class
    accuracy = accuracy_score(y_true, y_pred)

    return cm, sensitivity, specificity, accuracy

# Evaluate clf
cm_clf, sensitivity_clf, specificity_clf, accuracy_clf = evaluate_model(y_test, y_pred_clf)

# Evaluate clf2
cm_clf2, sensitivity_clf2, specificity_clf2, accuracy_clf2 = evaluate_model(y_test, y_pred_clf2)

# Display results
print("Model: clf (List Price only)")
print("Confusion Matrix:\n", cm_clf)
print(f"Sensitivity: {sensitivity_clf:.2f}, Specificity: {specificity_clf:.2f}, Accuracy: {accuracy_clf:.2f}\n")

print("Model: clf2 (NumPages, Thick, List Price)")
print("Confusion Matrix:\n", cm_clf2)
print(f"Sensitivity: {sensitivity_clf2:.2f}, Specificity: {specificity_clf2:.2f}, Accuracy: {accuracy_clf2:.2f}")

# Visualize confusion matrices
ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=['Paperback', 'Hardcover']).plot()
plt.title("Confusion Matrix: clf (List Price only)")
plt.show()

ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=['Paperback', 'Hardcover']).plot()
plt.title("Confusion Matrix: clf2 (NumPages, Thick, List Price)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Feature importances for clf2
importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

# Find the most important feature
most_important_feature_idx = np.argmax(importances)
most_important_feature = feature_names[most_important_feature_idx]

# Plot feature importances
plt.figure(figsize=(8, 6))
plt.barh(feature_names, importances, align="center")
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("Feature Importances for clf2")
plt.show()

# Report the most important feature
print(f"The most important predictor variable for clf2 is: {most_important_feature}")
```

Explanation of the .feature_importances_ Attribute
The .feature_importances_ attribute provides a normalized measure (summing to 1) of how much each feature contributes to reducing the impurity in the decision tree splits. A higher value indicates greater importance in determining the tree's predictions.

Q9

In linear regression, interpreting coefficients is straightforward: each coefficient represents the expected change in the target variable for a one-unit change in the predictor variable, holding all other variables constant. This direct relationship makes it easy to quantify the impact of each feature. In contrast, feature importances in decision trees are derived from how much each feature contributes to reducing impurity (e.g., Gini or entropy) across all splits in the tree. This indirect and aggregate measure reflects the feature's overall role in the tree's structure, but it doesn't specify the direction or magnitude of its effect on predictions.
